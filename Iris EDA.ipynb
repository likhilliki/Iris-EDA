```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Load iris dataset from seaborn
df = sns.load_dataset("iris")

# Display first 5 rows
df.head()
```

```python
# Dataset dimensions
print("Rows and Columns:", df.shape)

# Dataset information
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics
df.describe()
```

```python
sns.countplot(x="species", data=df)
plt.title("Distribution of Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
df.hist(bins=20, edgecolor="black")
plt.suptitle("Distribution of Iris Features")
plt.show()
```

```python
sns.pairplot(df, hue="species")
plt.show()
```

```python
for column in df.columns[:-1]:
    sns.boxplot(x="species", y=column, data=df)
    plt.title(f"{column} by Species")
    plt.show()
```

```python
# Correlation matrix (numerical columns only)
corr = df.drop(columns="species").corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.violinplot(x="species", y="petal_length", data=df)
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
###  Key Insights from EDA:
- Petal length and petal width are highly correlated
- Iris setosa is clearly distinguishable from other species
- Sepal features show more overlap among species
- Dataset is clean with no missing values
- Ideal dataset for classification problems
```
